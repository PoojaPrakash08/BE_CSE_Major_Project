```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
file_path = r"C:\\Users\\POOJA\\Desktop\\Major Project\\Datasets\\LSTM_ET_PM_2.csv"
df = pd.read_csv(file_path)

# Step 2: Plot the graphs
plt.figure(figsize=(10, 6))

# MaxTemp vs Et_PM
plt.subplot(2, 3, 1)
sns.scatterplot(data=df, x='MaxTemp', y='ET_PM')
plt.title('MaxTemp vs ET_PM')
plt.xlabel('MaxTemp')
plt.ylabel('ET_PM')

# MinTemp vs Et_PM
plt.subplot(2, 3, 2)
sns.scatterplot(data=df, x='MinTemp', y='ET_PM')
plt.title('MinTemp vs ET_PM')
plt.xlabel('MinTemp')
plt.ylabel('ET_PM')

# Rainfall vs Et_PM
plt.subplot(2, 3, 3)
sns.scatterplot(data=df, x='Rainfall', y='ET_PM')
plt.title('Rainfall vs ET_PM')
plt.xlabel('Rainfall')
plt.ylabel('ET_PM')

# Humidity vs Et_PM
plt.subplot(2, 3, 4)
sns.scatterplot(data=df, x='Humidity', y='ET_PM')
plt.title('Humidity vs ET_PM')
plt.xlabel('Humidity')
plt.ylabel('ET_PM')

# WindSpeed vs Et_PM
plt.subplot(2, 3, 5)
sns.scatterplot(data=df, x='WindSpeed', y='ET_PM')
plt.title('WindSpeed vs ET_PM')
plt.xlabel('WindSpeed')
plt.ylabel('ET_PM')

# Radiation vs Et_PM
plt.subplot(2, 3, 6)
sns.scatterplot(data=df, x='Radiation', y='ET_PM')
plt.title('Radiation vs ET_PM')
plt.xlabel('Radiation')
plt.ylabel('ET_PM')

plt.tight_layout()
plt.show()
```
