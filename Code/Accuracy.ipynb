```python
import matplotlib.pyplot as plt

# Accuracy values for three models
models = ['KNN', 'LSTM', 'ANN']
accuracy = [75.1073, 82.50, 79.7297]  # Accuracy values for Model A, B, and C respectively

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.bar(models, accuracy, color=['blue', 'green', 'red'])

# Adding title and labels
plt.title('Accuracy of Three Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')

# Showing the plot
plt.ylim(70.0, 85.0)  # Setting y-axis limits for better visualization
plt.show()
```
