```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\POOJA\\Desktop\\Major Project\\Datasets\\KNN_ET_PM.csv"

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Plotting
plt.figure(figsize=(12, 8))

# Plot max_temp vs ET_PM
plt.subplot(2, 3, 1)
sns.scatterplot(x='MaxTemp', y='ET_PM', data=data)
plt.title('Max Temperature vs ET_PM')

# Plot min_temp vs ET_PM
plt.subplot(2, 3, 2)
sns.scatterplot(x='MinTemp', y='ET_PM', data=data)
plt.title('Min Temperature vs ET_PM')

# Plot humidity vs ET_PM
plt.subplot(2, 3, 3)
sns.scatterplot(x='Humidity', y='ET_PM', data=data)
plt.title('Humidity vs ET_PM')

# Plot radiation vs ET_PM
plt.subplot(2, 3, 4)
sns.scatterplot(x='Radiation', y='ET_PM', data=data)
plt.title('Radiation vs ET_PM')

# Plot wind_speed vs ET_PM
plt.subplot(2, 3, 5)
sns.scatterplot(x='WindSpeed', y='ET_PM', data=data)
plt.title('Wind Speed vs ET_PM')

# Plot rainfall vs ET_PM
plt.subplot(2, 3, 6)
sns.scatterplot(x='Rainfall', y='ET_PM', data=data)
plt.title('Rainfall vs ET_PM')

plt.tight_layout()
plt.show()
```
